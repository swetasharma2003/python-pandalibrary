```python
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots(figsize=(14,5))
ax.plot([1,2,3,4],[1,4,2,3]);
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.scatter([2,6,7,8,9],[2,7,5,9,8]);
```

```python
fig,ax=plt.subplots(figsize=(12,10))
ax.bar([1,4,7,8],[2,7,9,5]);
```

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.bar([1,2,3,4,6,7],[3,4,7,9,6,9]);
ax.set_xlabel('x-axis');
ax.set_ylabel('y-axis');
ax.set_title('simple bar chart');
```

```python
fig,ax=plt.subplots()
ax.plot([2,3,4,5,10],[2,4,6,7,8],label='data1',color='red',marker='x',linestyle='dashed',markersize=8,linewidth=3);
ax.plot([1,6,8,9,10],[3,5,8,6,10],label='data2',color='green',marker='o',linestyle='dotted',markersize=8,linewidth=3);
ax.legend();
ax.set_xlabel('x-axis');
ax.set_ylabel('y-axis');
ax.set_title('simple plot');
```
